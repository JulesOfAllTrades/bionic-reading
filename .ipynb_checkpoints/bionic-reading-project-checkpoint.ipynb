# Bionic Reading

## What is Bionic Reading?
Bionic reading is the latest internet productivity hack. Basically, it bolds the first few letters of each word in a block of text. The theory is that it guides your eyes through the text faster. I can't speak to the validity of this--just think it is interesting.

## Project Plan

 - [x] Concatenate back into one string - don't forget to add spaces back in
 - [x] Display bionic text in a text cell using ipynb display HTML
 
### Bolding logic
 - If a word has only 1 letter, do not bold.  
 - If a word has 2 or 3 letters, bold the first letter.  
 - If word has 4 or 5 letters, bold the first two letters.  
 - If a word has 6 or more letters, bold the first three letters.

## The Function

```python
def bionic_text(txt):
    lines = txt.replace("\n", " fixlinebreak ")
    words = lines.split(" ")
    html_words = []
    for word in words:
        if word == "fixlinebreak":
            html_words.append(word)
        elif len(word) >=6 and word != "fixlinebreak": 
            bold_word = "<b>" + word[:3] + "</b>" + word[3:] + " "
            html_words.append(bold_word)
        elif len(word) >= 4:
            bold_word = "<b>" + word[:2] + "</b>" + word[2:] + " "
            html_words.append(bold_word)
        elif len(word) >= 2:
            bold_word = "<b>" + word[:1] + "</b>" + word[1:] + " "
            html_words.append(bold_word)
        elif len(word) >=0:
            html_words.append(word + " ")
    html_text = ""
    for hw in html_words:
        html_text += hw
    html_text_final = html_text.replace("fixlinebreak", " <br> ")
    return html_text_final
```
